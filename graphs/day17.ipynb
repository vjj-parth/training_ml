```python
import matplotlib.pyplot as plt
import numpy as np
```

# Types of graphs
1. plot -->         Line graph between two parameters
2. bar -->          bar graph of single parameter (optional categorical data)
3. hist -->         histogram of a bunch of data i.e. frequency of single parameter
4. scatter -->      dot graph between two parameters
5. boxplot / box --> statistical overview in form of box for a single parameter
                    mean, outliers, quantiles (25%, 75%), IQR
                    (optional - categorical data)
6. stem -->         more like bar graph, Lines instead of bars
7. violinplot -->   Violin shaped statistical overview graph
                    spead of data (more like hist)
                    single parameter (optional cat-data)

```python
x = np.linspace(0, 5, 11)
x1 = np.arange(1, 11, 3)
```

```python
y = x**3
```

```python
plt.plot(x,y)
plt.title('Plot Graph')
## plt.show()
plt.grid()
plt.xticks(x)
plt.yticks(range(0, 200,10))
plt.xlabel("x-->")
plt.ylabel("y-->")
```

```python
plt.bar(x, y)
plt.title('Bar Graph')
```

```python
plt.hist(y)
```

```python
plt.scatter(x,y)
plt.grid()
plt.xlabel("x-->")
plt.ylabel("y-->")
plt.xticks(x)
plt.yticks(range(0, 200,10))
```

```python
plt.boxplot(x)
```

```python
plt.stem(x, y)
```

```python
plt.violinplot(x)
```

## comparing two graphs

```python
y2 = x**4
```

```python
plt.plot(x, y)
plt.show()
plt.plot(x, y2)
plt.show()
```

```python
plt.plot(x, y, x, y2)
plt.legend(['y=x**3', 'y2=x**4'])
plt.show()
```

```python
fig = plt.figure(figsize = (10,5))
plt.plot(x, y, lw=3, marker='o', markerfacecolor = 'yellow')
plt.plot(x, y2, marker= 'v', markersize = 10)
plt.legend(['y=x**3', 'y2=x**4'])
plt.grid()
plt.show()
```
