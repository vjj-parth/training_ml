1. plot                     -- Simple Line Plot
2. plot.area                -- Filled Version of Line Plot
3. plot.bar  /  plot.barh   -- Bar Graph
4. plot.box                 -- Box Plot

6. plot.hexbin              -- Hexagon Version of Scatter with Density
7. plot.hist                -- Histogram
8. plot.kde                 -- Kernel Density Estimator (Density Graph)
9. plot.pie                 -- Pie Chart of Categorical Data

# Seaborn

```python
import seaborn as sns
```

## Categories
##### 1. Categorical Plots(hue can be used | comparison with categorical data) -->
    a. countplot
    b. barplot (one categorical , one numerical data, calculates mean by default)
    c. boxplot
    d. violinplot
    e. swarmplot
##### 2. Distribution Plots -->
    a. distplot
    b. displot
    c. jointplot
    d. pairplot
    e. kdeplot
    f. rugplot
    g. scatterplot
##### 3. Matrix Plots -->
    a. heatmap
    b. clustermap
    c. pairplot (same as above one)
##### 4. Regression Plots -->
    a. lmplot
    b. regplot
##### 5. Grid Plots -->
    a. PairGrid
    b. JointGrid
    c. FacetGrid

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe(include='all')
```

```python
tips['day'].value_counts()
```

```python
tips.groupby(['sex', 'time']) ['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day', data=tips)
```

```python
sns.countplot(x='day', data=tips, hue='sex')
```

# Bar Plot
One parameter must be categorical and one numerical

```python
sns.barplot(x='day', y='total_bill', data = tips)
```

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum)
```

```python
sns.barplot(x='day', y='total_bill', data=tips, hue='time', estimator=np.mean)
```

# Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='time')
```

# Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='time')
```

# Swarm Plot

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='time'); plt.show()
sns.swarmplot(x='day', y='total_bill', data=tips, hue='smoker'); plt.show()
```
